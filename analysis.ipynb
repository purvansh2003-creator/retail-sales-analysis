```python
import pandas as pd
import numpy as np
```

```python
# ENCODING TYPE MISMATCH 
# CSV files have different encoding 
# Kaggle Datasets are often : latin1 or cp1252
# Pandas assumes UTF-8 by default
superstore_csv = pd.read_csv('/home/purvansh-srivastava/Desktop/Python_for_Data_Analysis/mini_project_after_ch#5/Super_Store_Dataset/Sample - Superstore.csv',encoding='latin1')
# This works in mostly 90% cases
```

```python
superstore_csv.info()
```

```python
superstore_csv[superstore_csv['Country'] == 'United States' ]
```

```python
# Total Number of Orders
Total_Orders=superstore_csv['Order ID'].nunique()
print("Total Number of Orders = " ,Total_Orders)
```

```python
# Total Number of Sales
Sales=superstore_csv['Sales'].sum()
print("Total Number of Sales = $" ,Sales)
```

```python
# Total Profit
Total_profit=superstore_csv['Profit'].sum()
print("Total Profit = $" ,Total_profit)
```

```python
# Average Sales per Order

# First Calculate Total Sales per Order then find their mean
Total_sales_per_order = superstore_csv.groupby('Order ID')['Sales'].sum()
print('Average Sales per order = ',Total_sales_per_order.mean())
```

```python
# How many unique cities?
superstore_csv['City'].nunique()
```

```python
# Which region has the highest sales ?
Highest_sales_Regionwise = superstore_csv.groupby('Region')['Sales'].sum().max()
print('Region with highest sales :' ,superstore_csv.groupby('Region')['Sales'].sum().idxmax(),', Sales = ',Highest_sales_Regionwise)
#idxmax -> returns the index(label) of the maximum value
```

```python
# Which city has the highest profit ?
City_with_Highest_Profit = superstore_csv.groupby('City')['Profit'].sum().idxmax()
print('City with Highest Profit :',City_with_Highest_Profit)
```

```python
# Sales by category
superstore_csv.groupby('Category')['Sales'].sum().sort_values(ascending = False)
```

```python
# Profit by Sub-category
superstore_csv.groupby('Sub-Category')['Profit'].sum().sort_values(ascending = False)
```

```python
# Quantities Sold by Category
superstore_csv.groupby('Category')['Quantity'].sum().sort_values(ascending=False)
```

```python
# Which region is the most profitable ?
print('Most Profitable Region:',superstore_csv.groupby('Region')['Profit'].sum().idxmax())
```

```python
# Which category is the least profitable ?
print('Least Profitable Category:',superstore_csv.groupby('Category')['Profit'].sum().idxmin())
```

```python
# Are there any loss-making categories/sub-categories ?
cat_profit = superstore_csv.groupby('Category')['Profit'].sum()  
sub_cat_profit = superstore_csv.groupby('Sub-Category')['Profit'].sum() 
cat_profit[cat_profit<0]
sub_cat_profit[sub_cat_profit<0]
```

```python
# Top 5 city by sales
superstore_csv.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5)
```

```python
# All orders where Sales > 1000
orders = superstore_csv[['Order ID','Sales']]
orders[orders['Sales']>1000]
```

```python
# Orders where Profit < 0 (loss)
order_profit = superstore_csv.groupby('Order ID')['Profit'].sum() 
order_profit[order_profit < 0].sort_values(ascending=True)
```

```python
# Orders from a specific city
superstore_csv[superstore_csv['City'] == 'Los Angeles']
```

```python
# High Sales but low Quantity
# High Sales : how to define -> sales which are greater than average .
# low Quantity : how to define -> Quantity which are less than average .
High_Sales_low_Quantity = superstore_csv[
                             (superstore_csv['Sales'] > superstore_csv['Sales'].mean()) &
                              (superstore_csv['Quantity'] < superstore_csv['Quantity'].mean())  
                              ]  
High_Sales_low_Quantity
```

```python
# Create new Column -> Profit per item
superstore_csv['Profit_per_Item'] = superstore_csv['Profit'] / superstore_csv['Quantity'].replace(0,1)
superstore_csv
```

```python
# Product with highest Profit per item
idx=superstore_csv['Profit_per_Item'].idxmax()
superstore_csv.loc[idx]
```

```python
# Top 3 Most sold Products
superstore_csv.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(3)
```

```python
superstore_csv.groupby('Region')['Sales'].sum().plot(kind='bar')
```
